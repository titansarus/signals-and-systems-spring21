```python
nominator = [1,2,10]
denominator = [1,25,1.9*10**6,10]
```

<div style="direction:rtl">
سپس نماد‌ها و تابع $H$ را تعریف می‌کنیم. به نظر می‌رسد که در صورت سوال اشتباهی وجود دارد و توان دوم مخرج چاپ نشده است. تابعی که در این جا ترسیم کرده ایم، تابع زیر است:
$$H(j\omega) = \frac{(j\omega)^2 + 2 j\omega + 10}{(j\omega)^3 + 25 (j\omega)^2 + 1.9 \times 10^6 (j\omega) + 10}$$
</div>

```python
H = control.tf(nominator,denominator)
print(H)
```

```python
mag,phase,omega = control.bode(H)
```

<div style="direction:rtl">
حال بار دیگر بدون کتابخانه Control پیاده‌سازی می‌کنیم
</div>

```python
def H(w):
    nominator = (1j*w)**2 + 2j*w+10
    denominator = (1j*w)**3 + 25 * (1j*w)**2 + 1.9 * 10**6 * 1j*w + 10
    return nominator / denominator
```

<div style="direction:rtl">
سپس آن را رسم می‌کنیم.
</div>

```python
w = np.logspace(-6 , 4)
Mag_DB = 20*np.log10(np.abs(H(w)))
Mag_Normal = np.abs(H(w))
Phase = np.angle(H(w))*180/(np.pi)

fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax.set_xscale('log')
ax.set_xlabel("rad/sec")
ax.set_title("Magnitude (dB)")
ax.set_ylabel("Magnitude (dB)")
ax.plot(w , Mag_DB)

fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("rad/sec")
ax.set_ylabel("Magnitude")
ax.set_title("Magnitude")
ax.plot(w , Mag_Normal)



fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax.set_xscale('log')
ax.set_xlabel("rad/sec")
ax.set_ylabel("Degree")
ax.set_title("Phase")
ax.plot(w , Phase)

```

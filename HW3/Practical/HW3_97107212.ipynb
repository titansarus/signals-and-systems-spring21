<div style="direction:rtl">
ابتدا کتابخانه sympy و موارد لازم دیگر را ایمپورت می‌کنیم</div>

```python
import sympy
from sympy import re, im, I, E, symbols
from sympy.integrals.transforms import z
from sympy.integrals.transforms import inverse_laplace_transform
from sympy import *
init_printing(use_latex='mathjax')

from IPython.display import display

```

<div style="direction:rtl">
ابتدا ثوابت را تعریف می‌کنیم
</div>

```python
p1 = 0.7 + 0.5 *I
p2 = 0.6 - 0.3* I
z1 = 1 + 0*I
z2 = -1 + 0 * I
```

<div style="direction:rtl">
سپس نماد‌ها و تابع $H$ را تعریف می‌کنیم.
</div>

```python
z = symbols('z' , complex = True)
w = symbols('\omega' , real = True)
h = ((z - z1) * (z-z2))/((1 - p1 * (1/z) ) * (1 - p2 * (1/z)))
```

<div style="direction:rtl">
سپس تابع
    $H(e^{i \omega}$
    را تعریف کرده و اندازه و زاویه آن را بدست می‌آوریم
</div>

```python
result =  h.subs(z , E**(I * w))

amplitude= (sympy.Abs(result))
ang = (sympy.arg(result))
```

<div style="direction:rtl">
سپس آن را رسم می‌کنیم.
</div>

```python
sympy.plotting.plot(amplitude)
sympy.plotting.plot(angle)
```
